```python
import simpy
import random

def warehouse(env):
    while True:
        print(f"Order fulfilled at time {env.now}")
        yield env.timeout(random.randint(1, 3))  
env = simpy.Environment()
env.process(warehouse(env))
env.run(until=52)
```

```python
def supplier(env, warehouse):
    while True:
        yield env.timeout(random.randint(2, 5))  
        amount = random.randint(5, 10)
        warehouse.level += amount
        print(f"[{env.now}] Supplier delivered {amount} units. Stock: {warehouse.level}")

def customer(env, warehouse):
    while True:
        yield env.timeout(random.randint(3, 6))
        demand = random.randint(1, 4)
        if warehouse.level >= demand:
            warehouse.level -= demand
            print(f"[{env.now}] Customer bought {demand} units. Stock: {warehouse.level}")
        else:
            print(f"[{env.now}] Stockout! Customer wanted {demand} but only {warehouse.level} available.")

class Warehouse:
    def __init__(self):
        self.level = 20 

env = simpy.Environment()
warehouse = Warehouse()
env.process(supplier(env, warehouse))
env.process(customer(env, warehouse))
env.run(until=30)
```

```python
df['Order Date'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce')
df['Shipping Date'] = pd.to_datetime(df['shipping date (DateOrders)'], errors='coerce')
df['Order Quantity'] = pd.to_numeric(df['Order Item Quantity'], errors='coerce')
df['Product Price'] = pd.to_numeric(df['Product Price'], errors='coerce')

df_clean = df.dropna(subset=[
    'Order Date', 'Shipping Date', 'Order Quantity', 'Product Price'
])

df_clean['Pending Delivery'] = df_clean['Shipping Date'] > pd.Timestamp.now()
df_clean['Inventory Value'] = df_clean['Order Quantity'] * df_clean['Product Price']
df_clean['Order Fulfillment'] = df_clean['Pending Delivery'].apply(lambda x: 'Pending' if x else 'Delivered')
```

```python
daily_demand = df_clean.groupby(df_clean['Order Date'].dt.date)['Order Quantity'].sum()
safety_stock = 1.5 * daily_demand.mean()

summary = {
    'Total Orders': len(df_clean),
    'Total Pending Orders': df_clean['Pending Delivery'].sum(),
    'Total Delivered Orders': (~df_clean['Pending Delivery']).sum(),
    'Total Ordered Stock': df_clean['Order Quantity'].sum(),
    'Pending Stock': df_clean[df_clean['Pending Delivery']]['Order Quantity'].sum(),
    'Delivered Stock': df_clean[~df_clean['Pending Delivery']]['Order Quantity'].sum(),
    'Current Safety Stock Level': round(safety_stock)
}

summary_df = pd.DataFrame.from_dict(summary, orient='index', columns=['Value'])
summary_df.reset_index(inplace=True)
summary_df.columns = ['Metric', 'Value']
summary_df
```

```python
import simpy
import random

class Warehouse:
    def __init__(self, env, initial_stock, reorder_point, reorder_qty):
        self.env = env
        self.stock = initial_stock
        self.reorder_point = reorder_point
        self.reorder_qty = reorder_qty
        self.stockouts = 0
        self.action = env.process(self.run())

    def run(self):
        while True:
            if self.stock < self.reorder_point:
                print(f"[{self.env.now}] Reordering {self.reorder_qty} units")
                yield self.env.timeout(5)  # lead time
                self.stock += self.reorder_qty
                print(f"[{self.env.now}] Received stock. Current level: {self.stock}")
            yield self.env.timeout(1)

    def fulfill_order(self, quantity):
        if self.stock >= quantity:
            self.stock -= quantity
            print(f"[{self.env.now}] Order fulfilled: {quantity} units")
        else:
            print(f"[{self.env.now}] Stockout! Requested: {quantity}, Available: {self.stock}")
            self.stockouts += 1
            self.stock = 0

def customer_demand(env, warehouse):
    while True:
        yield env.timeout(random.randint(1, 3))
        quantity = random.randint(2, 8)
        warehouse.fulfill_order(quantity)

env = simpy.Environment()
warehouse = Warehouse(env, initial_stock=50, reorder_point=20, reorder_qty=30)
env.process(customer_demand(env, warehouse))
env.run(until=30)
```

```python
class DeliverySimulator:
    def __init__(self, env, orders_df):
        self.env = env
        self.orders = orders_df
        self.log = []

    def simulate_order(self, row):
        order_id = row['Order Id']
        delivery_time = row['Days for shipping (real)']
        profit = row['Order Profit Per Order']
        
        yield self.env.timeout(delivery_time)  # simulate shipping delay
        delivered = random.random() > 0.1  # 90% chance delivery success
        returned = not delivered or random.random() < 0.05  # 5% chance of return

        status = "Delivered"
        final_profit = profit

        if returned:
            status = "Returned"
            final_profit = -abs(profit)

        self.log.append({
            "Time": self.env.now,
            "Order Id": order_id,
            "Status": status,
            "Profit": final_profit,
            "Delivery Time": delivery_time
        })

    def run(self):
        for idx, row in self.orders.iterrows():
            self.env.process(self.simulate_order(row))
```

```python
df = pd.read_csv("DataCoSupplyChainDataset.csv", encoding="ISO-8859-1")
df['Days for shipping (real)'] = pd.to_numeric(df['Days for shipping (real)'], errors='coerce')
df['Order Profit Per Order'] = pd.to_numeric(df['Order Profit Per Order'], errors='coerce')
df = df.dropna(subset=['Days for shipping (real)', 'Order Profit Per Order', 'Order Id'])


sample_orders = df.sample(100).reset_index(drop=True)


env = simpy.Environment()
sim = DeliverySimulator(env, sample_orders)
sim.run()
env.run()


result_df = pd.DataFrame(sim.log)
result_df.head()
```
